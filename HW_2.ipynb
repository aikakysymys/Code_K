```python
from time import time
from sklearn.feature_extraction.text import CountVectorizer

def extract_words(texts):
    words = set()
    for text in texts:
        words = words.union(set(filter(lambda x: x != '', text.split(' '))))
    return list(words)

def count_vectorizer(texts):
    features_names = extract_words(texts)
    count_features = []
    for j in range(len(texts)):
        count = Counter(texts[j].split())
        local_counts = np.zeros(len(features_names))
        for i in range(len(features_names)):
            local_counts[i] = count.get(features_names[i], 0)
        count_features.append(local_counts)
    return features_names, np.array(count_features).T

s = time()
cv = CountVectorizer()
cv.fit_transform(df['text'].values)
print("CV time", time() - s)

s = time()
words, features = count_vectorizer(df['text'].values)
print("My time", time() - s)
```
